## Distribución t-student

```python
!pip install pingouin
import pingouin as pg

# Descarga de datos
url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/' +
       'Estadistica-machine-learning-python/master/data/births.csv')

datos = pd.read_csv(url, sep=',')
datos.head()
```

```python
# Filtrar la base de datos entre madres fumadoras y no fumadoras
peso_smokers = datos.loc[datos.smoke == 'smoker', 'weight']
peso_nonsmokers = datos.loc[datos.smoke == 'nonsmoker', 'weight']
peso_smokers
```

```python
# aplicando el t-test
pg.ttest(x = peso_smokers, y = peso_nonsmokers)
```

Como el $p-valor = 0.13>0.05$

$T = -1.4$ esta en la region de aceptacion

Es decir $μ_{1} = μ_{2}$

Por lo tanto, no hay diferencia entre los pesos de los bebés de madres fumadoras o no fumadoras.
