# Teorema del Límite Central (TLC)

Dada una muestra de datos $X_1, X_2, ..., X_n$ de media $μ$ y su varianza $\sigma^2$ (no importando la distribución), si $n$ es suficientemente grande entonces $\overline{X} \sim N(\mu, \frac{\sigma^2}{n})$.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Generaremos una normal desde una exponencial
Si cambiamos el tamaño de la muestra a numeros muy grandes, tiende a verse como una dist. normal.

Original:
- t_muestra = 5
- num_muestras = 2

```python
lambda_parametro = 0.5
t_muestra = 1000000 # num de elemento/variables por muestra
num_muestras = 1000 # num de muestras
muestras = np.random.exponential(scale = (1/lambda_parametro), size=(num_muestras, t_muestra))
muestras
```

```python
# Calcular el promedio por muestra
medias_muestrales = np.mean(muestras, axis = 1)
medias_muestrales
```

```python
# Crear un Datafreame para la media de las muestras
data = {'Muestra': range(1, num_muestras + 1), 'Media Muestral': medias_muestrales}
df = pd.DataFrame(data)
df
```

```python
# Crear un histograma con las medias muestrales
plt.hist(medias_muestrales, bins = 30, alpha = 0.6, color = 'b', label = 'Medias Muestrales')
plt.axvline(x = 1/lambda_parametro, color = 'r', linestyle = 'dashed', linewidth = 2, label = 'mu')
plt.xlabel('Valores de las medias muestrales')
plt.ylabel('Frecuencia de las medias muestrales (Probabilidad)')
plt.title('Teorema del Límite Central - Distribución Exponencial')
plt.legend()
plt.show()
```
